# Question 3.

```python
def convert_edge_list_to_adj_list(edges):
    adj_list = {}  # Initialize an empty dictionary for adjacency list
    for node1, node2 in edges:  # Loop through each edge (pair of nodes)
        if node1 not in adj_list:  # If node1 is not present in the adjacency list
            adj_list[node1] = []  # Initialize node1 with an empty list
        adj_list[node1].append(node2)  # Add node2 to the list of node1

        if node2 not in adj_list:  # If node2 is not present in the adjacency list
            adj_list[node2] = []  # Initialize node2 with an empty list
        adj_list[node2].append(node1)  # Add node1 to the list of node2

    return adj_list  # Return the adjacency list


def convert_matrix_to_adj_list(matrix):
    adj_dict = {}  # Initialize an empty dictionary for adjacency list
    for i in range(len(matrix)):  # Loop through each row in the matrix
        adj_dict[f"Node{i+1}"] = []  # Initialize an empty list for each node
        for j in range(len(matrix)):  # Loop through each column
            if matrix[i][j] == 1:  # If there is an edge between i and j
                adj_dict[f"Node{i+1}"].append(f"Node{j+1}")  # Add node to the list
                
    return adj_dict  # Return the adjacency list


def check_adjacency(graph, node_a, node_b):
    if isinstance(graph, list) and isinstance(graph[0], tuple):  # If the graph is in edge list format
        adj_graph = convert_edge_list_to_adj_list(graph)  # Convert edge list to adjacency list

        return node_a in adj_graph.get(node_b, [])  # Check if node_b has an edge to node_a
    
    elif isinstance(graph, list) and isinstance(graph[0], list):  # If the graph is in adjacency matrix format
        adj_graph = convert_matrix_to_adj_list(graph)  # Convert adjacency matrix to adjacency list

        return node_a in adj_graph.get(node_b, [])  # Check if node_b has an edge to node_a
    
    else:  # If the graph is already in adjacency list format
        return node_a in graph.get(node_b, [])  # Check if node_b has an edge to node_a


# Example graphs
edge_list_graph = [('Node1', 'Node2'), ('Node1', 'Node3'), ('Node1', 'Node5'), ('Node2', 'Node4'), ('Node3', 'Node5')]  # Edge list format

adj_matrix_graph = [
        [0, 1 , 1, 0, 1],  # Adjacency matrix format
        [1, 0 , 0, 1, 0],
        [1, 0 , 0, 0, 1],
        [0, 1 , 0, 0, 0],
        [1, 0 , 1, 0, 0]
    ]

adj_list_graph = {"Node1":["Node2", 'Node3', 'Node5'],  # Adjacency list format
                  "Node2":["Node1", 'Node4'],
                  "Node3":["Node1", 'Node5'],
                  "Node4":["Node2"],
                  'Node5':['Node1', 'Node3']
                  }

# Test the function
print(check_adjacency(edge_list_graph, "Node3", "Node1"))  # Test with edge list
```

# Question 5.

Check if a graph is connected.


```python
def convert_edge_list_to_adj_list(edges):  
    adj_list = {}
    for node1, node2 in edges:
        if node1 not in adj_list:
            adj_list[node1] = []
        adj_list[node1].append(node2)

        if node2 not in adj_list:
            adj_list[node2] = []
        adj_list[node2].append(node1)

    return adj_list

def convert_matrix_to_adj_list(matrix):
    adj_dict = {}
    for i in range(len(matrix)):
        adj_dict[f"Node{i+1}"] = []
        for j in range(len(matrix)):
            if matrix[i][j] == 1:
                adj_dict[f"Node{i+1}"].append(f"Node{j+1}")
                
    return adj_dict

def convert_graph(graph):
    if isinstance(graph, list) and isinstance(graph[0], tuple):
        return convert_edge_list_to_adj_list(graph)
    elif isinstance(graph, list) and isinstance(graph[0], list):
        return convert_matrix_to_adj_list(graph)
    else:
        return graph

def check_graph_connected(graph):
    graph = convert_graph(graph)
    
    nodes = list(graph.keys())  # List of nodes
    visited = []  # List of visited nodes
    queue = [nodes[0]]  # BFS starting from the first node

    while queue:
        node = queue.pop(0)  # Dequeue the first node
        if node not in visited:  # If the node hasn't been visited
            visited.append(node)  # Mark it as visited
            # Add unvisited neighbors to the queue
            for neighbor in graph[node]:
                if neighbor not in visited and neighbor not in queue:
                    queue.append(neighbor)

    # Check if all nodes are visited (graph is connected)
    return f"Is graph connected: {len(visited) == len(nodes)}"

# Example graphs
graph1 = [('Node1', 'Node2'), ('Node1', 'Node3'), ('Node2', 'Node3')]  # Edge list

graph2 = [  # Adjacency matrix
    [0, 1, 1, 0, 1],
    [1, 0, 0, 1, 0],
    [1, 0, 0, 0, 1],
    [0, 1, 0, 0, 0],
    [1, 0, 1, 0, 0]
]

graph3 = {  # Adjacency list
    "Node1": ["Node2", 'Node3', 'Node5'],
    "Node2": ["Node1", 'Node4'],
    "Node3": ["Node1", 'Node5'],
    "Node4": ["Node2"],
    'Node5': ['Node1', 'Node3']
}

print(check_graph_connected(graph1))  # Test with graph1
```

# Question 6.

Given a graph and a list of vertices, check if it forms a walk, or a trail or a path or none of these.

```python
def EL_to_AL(edge):  
    # Convert edge list to adjacency list
    AL = {}
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []  
        AL[node1].append(node2)  # Add node2 to node1's neighbors

        if node2 not in AL:
            AL[node2] = []  
        AL[node2].append(node1)  # Add node1 to node2's neighbors

    return AL

def AM_to_AL(Matrix):
    # Convert adjacency matrix to adjacency list
    AL = {}
    for i in range(len(Matrix)):
        AL[f"S{i+1}"] = []
        for j in range(len(Matrix)):
            if Matrix[i][j] == 1:
                AL[f"S{i+1}"].append(f"S{j+1}")
    return AL

def Graph_conversion(MyGraph):
    # Convert graph to adjacency list if not already in that format
    if isinstance(MyGraph, list) and isinstance(MyGraph[0], tuple):
        return EL_to_AL(MyGraph)  # Convert edge list to adjacency list
    elif isinstance(MyGraph, list) and isinstance(MyGraph[0], list):
        return AM_to_AL(MyGraph)  # Convert adjacency matrix to adjacency list
    return MyGraph  # Already in adjacency list format

def is_walk(graph, walk):
    # Check if the walk is valid
    for i in range(len(walk) - 1):
        if walk[i+1] not in graph[walk[i]]:
            return False  # Not connected
    return True  # All consecutive nodes are connected

def check_walk(walk):
    # Check if the walk is a Path, Trail, or neither
    repeated_nodes = []
    repeated_edges = []
    
    for node in walk:
        if walk.count(node) > 1 and node not in repeated_nodes:
            repeated_nodes.append(node)  # Track repeated nodes

    for i in range(len(walk) - 1):
        edge = (walk[i], walk[i + 1])  
        reverse_edge = (walk[i + 1], walk[i])

        if (edge in repeated_edges or reverse_edge in repeated_edges):
            continue  # Skip already counted edges

        count = walk.count(walk[i])  # Count edge occurrences
        if count > 1:
            repeated_edges.append(edge)  # Track repeated edges

    # Return walk type based on repetition of nodes or edges
    if repeated_edges:
        return "Not a Trail, Not a Path"  
    elif repeated_nodes:
        return "Trail"  
    return "Path"  

def analyze_walk(graph, walk):
    # Analyze if the walk is valid and its type (Path, Trail, etc.)
    graph = Graph_conversion(graph)  # Ensure the graph is in adjacency list format
    if is_walk(graph, walk):  # Check if the walk is valid
        print("It is a valid WALK.")
        return check_walk(walk)  # Return walk type (Path, Trail, etc.)
    return "Not a Valid Walk."

# Example graphs
MyGraph1 = [('S1', 'S2'), ('S1', "S3"), ('S1', 'S5'), ('S2', 'S4'), ('S3', 'S5')]  # Edge list format
MyGraph2 = [  # Adjacency matrix format
    [0, 1 , 1, 0, 1],
    [1, 0 , 0, 1, 0],
    [1, 0 , 0, 0, 1],
    [0, 1 , 0, 0, 0],
    [1, 0 , 1, 0, 0]
]
MyGraph3 = {  # Adjacency list format
    "S1": ["S2", 'S3', 'S5'],
    "S2": ["S1", 'S4'],
    "S3": ["S1", 'S5'],
    "S4": ["S2"],
    'S5': ['S1', 'S3']
} 

# Walk to analyze
given_walk = ["S1", "S3", "S5", "S1"]
print(analyze_walk(MyGraph2, given_walk))  # Analyze walk for MyGraph2
```

# Question 7.

Check if a given graph is a tree.

```python
def EL_to_AL(edge):  
    # Convert edge list to adjacency list
    AL = {}
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []  # Initialize the node if not already in AL
        AL[node1].append(node2)  # Add the second node to the first node's adjacency list

        if node2 not in AL:
            AL[node2] = []  # Initialize the node if not already in AL
        AL[node2].append(node1)  # Add the first node to the second node's adjacency list
    return AL

def AM_to_AL(Matrix):
    # Convert adjacency matrix to adjacency list
    AL = {}
    for i in range(len(Matrix)):
        AL[f"S{i+1}"] = []  # Create an empty list for each node
        for j in range(len(Matrix)):
            if Matrix[i][j] == 1:  # If there's an edge between nodes i and j
                AL[f"S{i+1}"].append(f"S{j+1}")  # Add a neighboring node
    return AL

def Graph_conversion(graph):
    # Convert graph to adjacency list format if it's not already
    if isinstance(graph, list) and isinstance(graph[0], tuple):
        return EL_to_AL(graph)  # Edge list to adjacency list
    elif isinstance(graph, list) and isinstance(graph[0], list):
        return AM_to_AL(graph)  # Adjacency matrix to adjacency list
    return graph  # If already an adjacency list, return as is

def is_graph_connected(graph):
    # Check if the graph is connected using BFS
    graph = Graph_conversion(graph)
    nodes = list(graph.keys())
    visited = set()
    queue = [nodes[0]]  # Start BFS from the first node

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

    # Check if all nodes were visited
    return len(visited) == len(nodes)

def has_cycle(graph, node, visited, parent):
    # Helper function to detect a cycle using DFS
    visited.add(node)
    for neighbor in graph[node]:
        if neighbor not in visited:
            if has_cycle(graph, neighbor, visited, node):  # Recursively check neighbors
                return True
        elif neighbor != parent:  # If a neighbor is visited and it's not the parent, there's a cycle
            return True
    return False

def is_tree(graph):
    # Check if the graph is a tree
    if not is_graph_connected(graph):  # A tree must be connected
        return "Graph is not connected, so it's not a tree."

    visited = set()
    start_node = list(graph.keys())[0]  # Start DFS from any node
    if has_cycle(graph, start_node, visited, None):  # Check for cycles
        return "Graph has a cycle, so it's not a tree."

    edge_count = sum(len(neighbors) for neighbors in graph.values()) // 2  # Total edges
    node_count = len(graph)  # Total nodes

    # A tree must have exactly n-1 edges for n nodes
    if edge_count != node_count - 1:
        return "Graph doesn't have n-1 edges, so it's not a tree."

    return "Graph is a Tree"  # All conditions met, it's a tree

# Sample graph for testing
MyGraph4 = [('S1', 'S2'), ('S2', 'S3'), ('S3', 'S4')]  # This is a tree

print(is_tree(MyGraph4))  # Check if MyGraph4 is a tree
```

# Question 8.
Given a connected cyclic graph, find its spanning tree.

```python
def EL_to_AL(edge):  
    # Convert edge list to adjacency list
    AL = {}
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []  # Initialize node1 if not in AL
        AL[node1].append(node2)  # Add node2 as a neighbor of node1

        if node2 not in AL:
            AL[node2] = []  # Initialize node2 if not in AL
        AL[node2].append(node1)  # Add node1 as a neighbor of node2
    return AL

def AM_to_AL(matrix):
    # Convert adjacency matrix to adjacency list
    AL = {}
    for i in range(len(matrix)):
        AL[f"S{i+1}"] = []  # Create an empty list for each node
        for j in range(len(matrix)):
            if matrix[i][j] == 1:  # If there's an edge between nodes i and j
                AL[f"S{i+1}"].append(f"S{j+1}")  # Add a neighboring node
    return AL

def Graph_conversion(graph):
    # Convert graph to adjacency list format if it's not already
    if isinstance(graph, list) and isinstance(graph[0], tuple):
        return EL_to_AL(graph)  # Edge list to adjacency list
    elif isinstance(graph, list) and isinstance(graph[0], list):
        return AM_to_AL(graph)  # Adjacency matrix to adjacency list
    return graph  # If already in adjacency list, return as is

def find_spanning_tree(graph):
    # Find a spanning tree using DFS
    graph = Graph_conversion(graph)
    visited = set()  # Set to track visited nodes
    tree_edges = []  # List to store edges of the spanning tree

    def dfs(node):
        visited.add(node)  # Mark the node as visited
        for neighbor in graph[node]:  # Visit all unvisited neighbors
            if neighbor not in visited:
                tree_edges.append((node, neighbor))  # Add edge to tree
                dfs(neighbor)  # Recursively visit neighbors

    # Start DFS from the first node
    start_node = list(graph.keys())[0]
    dfs(start_node)

    return tree_edges

# Test Graphs
MyGraph1 = [('S1', 'S2'), ('S1', 'S3'), ('S2', 'S3')]  # Edge List
MyGraph2 = [  # Adjacency Matrix
    [0, 1 , 1, 0, 1],
    [1, 0 , 0, 1, 0],
    [1, 0 , 0, 0, 1],
    [0, 1 , 0, 0, 0],
    [1, 0 , 1, 0, 0]
]
MyGraph3 = {  # Adjacency List
    "S1": ["S2", "S3", "S5"],
    "S2": ["S1", "S4"],
    "S3": ["S1", "S5"],
    "S4": ["S2"],
    "S5": ["S1", "S3"]
}

# Run Tests
print("Spanning Tree of MyGraph1 (Edge List):", find_spanning_tree(MyGraph1))
print("Spanning Tree of MyGraph2 (Adj Matrix):", find_spanning_tree(MyGraph2))
print("Spanning Tree of MyGraph3 (Adj List):", find_spanning_tree(MyGraph3))
```

# Question 10.
Given a tree, check if it's a binary tree.

```python
def EL_to_AL(edge):  
    # Convert edge list to adjacency list
    AL = {}
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []
        AL[node1].append(node2)
        
        if node2 not in AL:
            AL[node2] = []
        AL[node2].append(node1)
    return AL

def AM_to_AL(matrix):
    # Convert adjacency matrix to adjacency list
    AL = {}
    for i in range(len(matrix)):
        AL[f"S{i+1}"] = []
        for j in range(len(matrix)):
            if matrix[i][j] == 1:
                AL[f"S{i+1}"].append(f"S{j+1}")
    return AL

def Graph_conversion(graph):
    # Convert graph to adjacency list format if it's in edge list or matrix form
    if isinstance(graph, list) and isinstance(graph[0], tuple):
        return EL_to_AL(graph)  # Edge list to adjacency list
    elif isinstance(graph, list) and isinstance(graph[0], list):
        return AM_to_AL(graph)  # Matrix to adjacency list
    return graph  # Already in adjacency list

def is_binary_tree(graph):
    # Check if the graph is a binary tree (each node can have at most 2 neighbors)
    graph = Graph_conversion(graph)
    for node in graph:
        if len(graph[node]) > 2:  # Binary tree condition
            return False
    return True

# Test Graphs
MyGraph1 = {  # Adjacency List
    "S1": ["S2", "S3", "S5"],
    "S2": ["S1", "S4"],
    "S3": ["S1", "S5"],
    "S4": ["S2"],
    "S5": ["S1", "S3"]
}

MyGraph2 = [  # Edge List
    ('S1', 'S2'), ('S1', 'S3'), ('S1', 'S5'), ('S2', 'S4'), ('S3', 'S5')
]

MyGraph3 = [  # Adjacency Matrix
    [0, 1, 1, 0, 1],
    [1, 0, 0, 1, 0],
    [1, 0, 0, 0, 1],
    [0, 1, 0, 0, 0],
    [1, 0, 1, 0, 0]
]

# Run Tests
print("Is MyGraph1 a binary tree?", is_binary_tree(MyGraph1))  # Should return True or False
print("Is MyGraph2 a binary tree?", is_binary_tree(MyGraph2))  # Should return True or False
print("Is MyGraph3 a binary tree?", is_binary_tree(MyGraph3))  # Should return True or False
```

# Question 11.
Given a tree and a node, find its height.

```python
# Convert Edge List to Adjacency List
def EL_to_AL(edge):  
    AL = {}
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []
        AL[node1].append(node2)
        
        if node2 not in AL:
            AL[node2] = []
        AL[node2].append(node1)
    return AL

# Convert Adjacency Matrix to Adjacency List
def AM_to_AL(Matrix):
    AL = {}
    for i in range(len(Matrix)):
        AL[f"S{i+1}"] = []
        for j in range(len(Matrix)):
            if Matrix[i][j] == 1:
                AL[f"S{i+1}"].append(f"S{j+1}")
    return AL

# General graph conversion function
def Graph_conversion(MyGraph):
    if isinstance(MyGraph, list) and isinstance(MyGraph[0], tuple):
        return EL_to_AL(MyGraph)  # Convert edge list to adjacency list
    elif isinstance(MyGraph, list) and isinstance(MyGraph[0], list):
        return AM_to_AL(MyGraph)  # Convert adjacency matrix to adjacency list
    return MyGraph  # Already in adjacency list form

# Recursive function to find the height of a node
def find_node_height(graph, node, visited):
    visited.add(node)  # Mark node as visited
    children = graph[node]  # Get the children (neighbors)

    if not children:  # If no children, it is a leaf node
        return 0

    max_height = 0
    for child in children:
        if child not in visited:
            child_height = find_node_height(graph, child, visited)
            max_height = max(max_height, child_height)

    return max_height + 1  # Height of the node = max child height + 1

# Function to find the height of a specific node
def get_height_of_node(MyGraph, node):
    graph = Graph_conversion(MyGraph)
    visited = set()  # To track visited nodes
    return find_node_height(graph, node, visited)

# Test Graphs
MyGraph1 = [('S1', 'S2'), ('S1', 'S3'), ('S2', 'S4')]  # Edge List
MyGraph2 = {  # Adjacency List
    "S1": ["S2", "S3"],
    "S2": ["S1", "S4"],
    "S3": ["S1"],
    "S4": ["S2"]
}

# Test: Find height of node 'S2' in MyGraph1 and 'S1' in MyGraph2
print("Height of node S2 in MyGraph1:", get_height_of_node(MyGraph1, 'S2'))
print("Height of node S1 in MyGraph2:", get_height_of_node(MyGraph2, 'S1'))
```

# Question 12.
Given a tree and a node, find its depth.

```python
# Convert Edge List to Adjacency List
def EL_to_AL(edge):  
    AL = {}
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []
        AL[node1].append(node2)
        
        if node2 not in AL:
            AL[node2] = []
        AL[node2].append(node1)
    return AL

# Convert Adjacency Matrix to Adjacency List
def AM_to_AL(Matrix):
    AL = {}
    for i in range(len(Matrix)):
        AL[f"S{i+1}"] = []
        for j in range(len(Matrix)):
            if Matrix[i][j] == 1:
                AL[f"S{i+1}"].append(f"S{j+1}")
    return AL

# General graph conversion function
def Graph_conversion(MyGraph):
    if isinstance(MyGraph, list) and isinstance(MyGraph[0], tuple):
        return EL_to_AL(MyGraph)  # Convert edge list to adjacency list
    elif isinstance(MyGraph, list) and isinstance(MyGraph[0], list):
        return AM_to_AL(MyGraph)  # Convert adjacency matrix to adjacency list
    return MyGraph  # Already in adjacency list form

# Recursive function to find the depth of a node
def find_node_depth(graph, node, target, depth, visited):
    visited.add(node)  # Mark node as visited
    if node == target:  # If we found the target node
        return depth

    for neighbor in graph[node]:
        if neighbor not in visited:
            result = find_node_depth(graph, neighbor, target, depth + 1, visited)
            if result != -1:  # If target is found
                return result
    return -1  # If target is not found

# Function to find the depth of a specific node
def get_depth_of_node(MyGraph, target):
    graph = Graph_conversion(MyGraph)  # Convert to adjacency list
    visited = set()  # Set to track visited nodes
    root_node = list(graph.keys())[0]  # Start from the first node
    return find_node_depth(graph, root_node, target, 0, visited)

# Test Graphs
MyGraph1 = [('S1', 'S2'), ('S1', 'S3'), ('S2', 'S4')]  # Edge List
MyGraph2 = {  # Adjacency List
    "S1": ["S2", "S3"],
    "S2": ["S1", "S4"],
    "S3": ["S1"],
    "S4": ["S2"]
}

# Test: Find depth of node 'S4' in MyGraph1 and MyGraph2
print("Depth of node S4 in MyGraph1:", get_depth_of_node(MyGraph1, 'S4'))
print("Depth of node S4 in MyGraph2:", get_depth_of_node(MyGraph2, 'S4'))
```
